```python
import numpy as np
import pandas as pd
from library.constants import *
import matplotlib.pyplot as plt

color=["#332288","#117733","#44AA99","#88CCEE","#DDCC77","#CC6677"]
```

```python
#Sensitivity(Recall)
rf_old=pd.read_csv("./logs/exp/overall_rf_old_Delhi.csv")['rec_score_test']
rf=pd.read_csv("./logs/exp/overall_rf_Delhi.csv")['rec_score_test']
rnn=pd.read_csv("./logs/exp/overall_rnn_Delhi.csv")['rec_score_test']

fig, ax = plt.subplots(figsize=figsize)
#RF OLD
ax.bar(x=np.arange(len(rf_old))-0.2,height=rf_old,align='center', alpha=1, ecolor='k',
       capsize=10,color=color[0],
       width=0.2,ec='k',linewidth=linewidth-2,
       error_kw =dict(linestyle='-',linewidth=3.0, color='black',markeredgewidth=3),label="RF")

#RF
ax.bar(x=np.arange(len(rf)),height=rf,align='center', alpha=1, ecolor='k',
       capsize=10,color=color[1],
       width=0.2,ec='k',linewidth=linewidth-2,#hatch='oo',
       error_kw =dict(linestyle='-',linewidth=3.0, color='black',markeredgewidth=3),label="RF with temporal feat.")
#RNN
ax.bar(x=np.arange(len(rnn))+0.2,height=rnn,align='center', alpha=1, ecolor='k',
       capsize=10,color=color[2],
       width=0.2,ec='k',linewidth=linewidth-2,#hatch='//',
       error_kw =dict(linestyle='-',linewidth=3.0, color='black',markeredgewidth=3),label="Revised model")

ax.set_ylim(0.55,0.85)

ax.set_xlabel("Device",fontsize=fontsize, fontweight='bold')
ax.set_ylabel("Sensitivity",fontsize=fontsize, fontweight='bold')

ax.set_xticks(np.arange(len(rf_old)))
ax.set_xticklabels([f"AQMS {i}" for i in range(len(rf_old))],rotation=30,fontsize=fontsize, fontweight='bold')
plt.yticks(fontsize=fontsize, fontweight='bold')
plt.legend(loc='upper left',prop=dict(size=fontsize-2,weight="bold"))

plt.tight_layout()
plt.grid()
fig.savefig("./logs/figs/delhi_sensi.png")
```
