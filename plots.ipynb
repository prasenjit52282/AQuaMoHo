```python
import numpy as np
import pandas as pd
import seaborn as sns
from library.constants import *
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

color=sns.color_palette()#["#332288","#117733","#44AA99","#88CCEE","#DDCC77","#CC6677"]
```

```python
#Sensitivity(Recall)
rf_old=pd.read_csv("./logs/exp/overall_rf_old_Delhi.csv")['rec_score_test']
rf=pd.read_csv("./logs/exp/overall_rf_Delhi.csv")['rec_score_test']
rnn=pd.read_csv("./logs/exp/overall_rnn_Delhi.csv")['rec_score_test']

fig, ax = plt.subplots(figsize=(7,5))
#RF OLD
ax.bar(x=np.arange(len(rf_old))-0.2,height=rf_old,align='center', alpha=1, ecolor='k',
       capsize=10,color=color[0],
       width=0.2,ec='k',linewidth=1,
       error_kw =dict(linestyle='-',linewidth=3.0, color='black',markeredgewidth=3),label="RF")

#RF
ax.bar(x=np.arange(len(rf)),height=rf,align='center', alpha=1, ecolor='k',
       capsize=10,color=color[1],
       width=0.2,ec='k',linewidth=1,#hatch='oo',
       error_kw =dict(linestyle='-',linewidth=3.0, color='black',markeredgewidth=3),label="RF with temporal feat.")
#RNN
ax.bar(x=np.arange(len(rnn))+0.2,height=rnn,align='center', alpha=1, ecolor='k',
       capsize=10,color=color[2],
       width=0.2,ec='k',linewidth=1,#hatch='//',
       error_kw =dict(linestyle='-',linewidth=3.0, color='black',markeredgewidth=3),label="LSTM-based model")

ax.set_ylim(0.55,0.85)

ax.set_xlabel("Device",fontsize=18)
ax.set_ylabel("Sensitivity",fontsize=18)

ax.set_xticks(np.arange(len(rf_old)))
ax.set_xticklabels([f"AQMS {i}" for i in range(len(rf_old))],rotation=30,fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='upper left',prop=dict(size=18))

plt.tight_layout()
plt.grid()
fig.savefig("./logs/figs/delhi_sensi.png")
```

```python
#Specificity (new)
rf_old=pd.read_csv("./logs/exp/overall_rf_old_Delhi.csv")['spec_score_test']
rf=pd.read_csv("./logs/exp/overall_rf_Delhi.csv")['spec_score_test']
rnn=pd.read_csv("./logs/exp/overall_rnn_Delhi.csv")['spec_score_test']

fig, ax = plt.subplots(figsize=(7,5))
#RF OLD
ax.bar(x=np.arange(len(rf_old))-0.2,height=rf_old,align='center', alpha=1, ecolor='k',
       capsize=10,color=color[0],
       width=0.2,ec='k',linewidth=1,
       error_kw =dict(linestyle='-',linewidth=3.0, color='black',markeredgewidth=3),label="RF")

#RF
ax.bar(x=np.arange(len(rf)),height=rf,align='center', alpha=1, ecolor='k',
       capsize=10,color=color[1],
       width=0.2,ec='k',linewidth=1,#hatch='oo',
       error_kw =dict(linestyle='-',linewidth=3.0, color='black',markeredgewidth=3),label="RF with temporal feat.")
#RNN
ax.bar(x=np.arange(len(rnn))+0.2,height=rnn,align='center', alpha=1, ecolor='k',
       capsize=10,color=color[2],
       width=0.2,ec='k',linewidth=1,#hatch='//',
       error_kw =dict(linestyle='-',linewidth=3.0, color='black',markeredgewidth=3),label="LSTM-based model")

ax.set_ylim(0.85,1)

ax.set_xlabel("Device",fontsize=18)
ax.set_ylabel("Specificity",fontsize=18)

ax.set_xticks(np.arange(len(rf_old)))
ax.set_xticklabels([f"AQMS {i}" for i in range(len(rf_old))],rotation=30,fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='upper left',prop=dict(size=18))

plt.tight_layout()
plt.grid()
fig.savefig("./logs/figs/delhi_speci.png")
```
